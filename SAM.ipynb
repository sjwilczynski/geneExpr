```r
library(samr)
library(dplyr)
```

```r
microarray <- t(as.matrix(read.csv(file="./data/microarray_train.csv", header=TRUE, sep=",")))
labels <- read.csv(file="./data/labels_train.csv", header=TRUE, sep=",")
labels = labels + 1
```

```r
genes <- read.csv(file="./data/microarray_data.csv", sep = ",", header = T)
```

```r
gene_names = as.character(unlist(genes %>% select(GeneSymbol)))
```

```r
data = list(x=microarray, y=labels, geneid=(1:length(gene_names)), genenames=gene_names, logged2=T)
```

```r
result <- samr(data = data, resp.type = "Two class unpaired", assay.type = "array", random.seed = 42)
```

```r
delta.table <- samr.compute.delta.table(result)
delta.table
```

```r
delta=2.46
samr.plot(result, delta)
```

```r
siggenes.table <-samr.compute.siggenes.table(result, delta, data, delta.table)
```

```r
siggenes.table
```

```r
significant.genes <- c(siggenes.table$genes.up[, "Gene ID"], siggenes.table$genes.lo[, "Gene ID"])
length(significant.genes)
```

```r
print(significant.genes[1:10])
```

```r
for (delta in c(2.0, 2.14, 2.30, 2.46, 2.62, 2.8, 2.97, 3.16, 3.34, 3.54, 3.74, 3.94, 4.15)) {
    siggenes.table <-samr.compute.siggenes.table(result, delta, data, delta.table)
    significant.genes <- c(siggenes.table$genes.up[, "Gene ID"], siggenes.table$genes.lo[, "Gene ID"])
    write.csv(matrix(significant.genes, nrow=1), file=paste0('./sam_results/sig_genes_', length(significant.genes), '.csv'), row.names=FALSE)
}
```
