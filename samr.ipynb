```r
library(samr)
```

```r
microarray <- t(as.matrix(read.csv(file="./data/microarray_train.csv", header=TRUE, sep=",")))
labels <- read.csv(file="./data/labels_train.csv", header=TRUE, sep=",")
labels = labels + 1
```

```r
dim(microarray)
length(labels)
```

```r
data = list(x=microarray, y=labels, logged2=T)
```

```r
result <- samr(data = data, resp.type = "Two class unpaired", assay.type = "array", random.seed = )
```

```r
delta=.4
samr.plot(result,delta)
```

```r
delta.table <- samr.compute.delta.table(result)
```

```r
siggenes.table <-samr.compute.siggenes.table(result, 2.46, data, delta.table)
```
